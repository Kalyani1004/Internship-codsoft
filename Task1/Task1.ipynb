```python
from tkinter import *

def add_item(entry, listbox, checkboxes):
    new_task = entry.get()
    if new_task:
        listbox.insert(END, new_task)
        checkboxes.append(IntVar())
        Checkbutton(tasks_frame, variable=checkboxes[-1], bg="Lightblue").pack(anchor=W)
        entry.delete(0, END)
        save_tasks_to_file(listbox.get(0, END))

def delete_item(listbox, checkboxes):
    try:
        for i in range(len(checkboxes) - 1, -1, -1):
            if checkboxes[i].get() == 1:  
                listbox.delete(i)
                for widget in tasks_frame.winfo_children():
                    if isinstance(widget, Checkbutton) and widget['variable'] == checkboxes[i]:
                        widget.destroy()
                del checkboxes[i]
        save_tasks_to_file(listbox.get(0, END))
    except IndexError:
        pass



def save_tasks_to_file(tasks):
    with open('D:/MCA/Internship-codsoft/Tasks.txt', 'w') as tasks_file:
        for task in tasks:
            tasks_file.write(task + '\n')

root = Tk()
root.title('To-Do List of Kalyani')
root.geometry('350x450')
root.resizable(0, 0)
root.config(bg="Lightblue")

Label(root, text='To Do List of Kalyani', bg="Lightblue", font=("Bell MT", 20)).pack(pady=10)

tasks_frame = Frame(root, bg="Lightblue")
tasks_frame.pack(pady=10)

tasks_scrollbar = Scrollbar(tasks_frame)
tasks_scrollbar.pack(side=RIGHT, fill=Y)

tasks_listbox = Listbox(tasks_frame, selectbackground='Gold', bg='White', font=('Helvetica', 12), height=12, width=25, yscrollcommand=tasks_scrollbar.set)
tasks_listbox.pack(side=LEFT, fill=BOTH)

tasks_scrollbar.config(command=tasks_listbox.yview)

checkboxes = []

with open('D:/MCA/Internship-codsoft/Tasks.txt', 'r+') as tasks_file:
    for task in tasks_file:
        tasks_listbox.insert(END, task.strip())
        checkboxes.append(IntVar())
        Checkbutton(tasks_frame, variable=checkboxes[-1], bg="Lightblue").pack(anchor=W)

new_item_entry = Entry(root, width=30)
new_item_entry.pack(pady=10)

add_btn = Button(root, text='Add Item', bg='White', width=10, font=('Helvetica', 12), command=lambda: add_item(new_item_entry, tasks_listbox, checkboxes))
add_btn.pack(pady=5)

delete_btn = Button(root, text='Delete Item', bg='White', width=10, font=('Helvetica', 12), command=lambda: delete_item(tasks_listbox, checkboxes))
delete_btn.pack(pady=5)

root.mainloop()
```

```python
def add_item(entry: Entry, listbox: Listbox):
    new_task = entry.get()

    listbox.insert(END, new_task)

    with open('tasks.txt', 'a') as tasks_list_file:
        tasks_list_file.write(f'{new_task}\n')


def delete_item(listbox, checkboxes):
    try:
        # Iterate in reverse order to avoid index shifting when deleting items
        for i in range(len(checkboxes) - 1, -1, -1):
            if checkboxes[i].get() == 1:  # If checkbox is checked
                listbox.delete(i)
                del checkboxes[i]

        # Rewrite tasks to the file after deleting checked tasks
        with open('tasks.txt', 'w') as tasks_list_file:
            for i in range(listbox.size()):
                tasks_list_file.write(listbox.get(i) + '\n')

    except IndexError:
        pass
```
